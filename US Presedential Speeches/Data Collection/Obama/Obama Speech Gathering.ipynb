### Create function for extracting text data from a pdf file

A lot of the speeches we import are going to be extracted from pdf files. Therefore we'll create a fucntion that can extract text from a pdf file. <br>

The source for the speeches in pdf format is the following website: https://www.americanrhetoric.com/speeches

```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ""
            for page_number in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_number]
                text += page.extract_text()
            return text
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

## Transcript Sentence Tokenization

We'll create some functions that cleanse our speech transcripts, then perform some **Tokenization** . Tokenzation is a process of splitting text into meaningful segments. In our case, to conduct the Sentiment Analysis, we want to split our speech transcripts into sentences. We'll then assess the sentiment for each of these sentences in a separate notebook. <br>

We'll test the logic out first on one of the transcripts, then create a function once the logic is finalised and apply it on ~the other transcriprts.

##### Sentence Tokenization using Spacy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('sentencizer')
nlp.pipe_names
```

```python
spacy_sen_tokens = []
doc = nlp(BO_Feb2009)
for sentence in doc.sents:
    spacy_sen_tokens.append(sentence)

for each in spacy_sen_tokens[:20]:
    print(f"sentence: {each}")
```

##### Sentence Tokenization using NLTK

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
nltk_sen_list = sent_tokenize(BO_Feb2009)
for each in nltk_sen_list[:20]:
    print(f"sentence: {each}")
```

Review: both the spacy sentencizer and nltk sentencizer are fairly comprehensive. There appear to be some short sentences but on the whole they both did a fairly good job of splitting the transcript out into understandable and sensible sentences. We'll adopt the spacy sentencizer to perform this task, purely on the basis that we may want to perform other nlp practices later on such as POS and NER, and we can use the Spacy pipeline we've initiated to keep everything consistent.

### Apply Sentencization to all Transcripts 

```python
def transcript_transform(transcript):
    # convert transcript string to nlp doc
    doc = nlp(transcript)
    list_of_tokens = []
    # apply sentence tokenization
    for sentence in doc.sents:
        text = re.sub('\n', '', str(sentence))
        list_of_tokens.append(str(text))
    # convert to dataframe
    token_df = pd.DataFrame(data=list_of_tokens, columns=['sentence'])
    return token_df
```
