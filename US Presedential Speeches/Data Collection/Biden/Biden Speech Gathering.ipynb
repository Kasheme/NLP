# Biden Speech Gathering

## Data Collection

```python
import functions as fnc
import requests
from bs4 import BeautifulSoup
```

### Biden - Speech on the Drawdown of US Forces in Afghanistan: 8th July 2021

```python
transcript_url = "https://millercenter.org/the-presidency/presidential-speeches/july-8-2021-speech-drawdown-us-forces-afghanistan"
response = requests.get(transcript_url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
target_div = soup.find('div', class_='transcript-inner')
if target_div:
    text_inside_div = str(target_div.text)
print(text_inside_div)
```

```python
start_substring = "Good afternoon. Earlier today, I was briefed"
end_substring = "May God bless you all, and may God protect our troops"

end_index_startsub = text_inside_div.find(start_substring)
end_index_speech = text_inside_div.find(end_substring) + len(end_substring)

print(end_index_startsub)
print(end_index_speech)
```

```python
transcript = text_inside_div[27:11882]
JB_Jul2021 = transcript
```

### Biden - Remarks on Situation in Afghansitan: 16th August 2021

```python
transcript_url = "https://millercenter.org/the-presidency/presidential-speeches/august-16-2021-remarks-situation-afghanistan"
response = requests.get(transcript_url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
target_div = soup.find('div', class_="transcript-inner")
if target_div:
    text_inside_div = str(target_div.text)
print(text_inside_div)
```

```python
text_inside_div = text_inside_div.replace("Transcript","")
text_inside_div = text_inside_div.replace("THE PRESIDENT:","") 
```

```python
JB_Aug2021 = text_inside_div
```

### Biden - Statement on Terror Attacks in Afghanistan: 26th August 2021

```python
transcript_url = "https://millercenter.org/the-presidency/presidential-speeches/august-26-2021-statement-terror-attacks-afghanistan"
response = requests.get(transcript_url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
target_div = soup.find('div', class_="transcript-inner")
if target_div:
    text_inside_div = str(target_div.text)
print(text_inside_div)
```

```python
start_substring = "Been a tough day."
end_substring = "We will find them and we will get them out."

end_index_startsub = text_inside_div.find(start_substring)
end_index_speech = text_inside_div.find(end_substring) + len(end_substring)

print(end_index_startsub)
print(end_index_speech)
```

```python
transcript = text_inside_div[26:6603]
transcript = transcript.replace("(A moment of silence is taken.)","")
JB_Aug2021_2 = transcript
```

## Sentence Tokenization

```python
JB_Jul2021_df = fnc.transcript_transform(JB_Jul2021)
JB_Aug2021_df = fnc.transcript_transform(JB_Aug2021)
JB_Aug2021_2_df = fnc.transcript_transform(JB_Aug2021_2)
```

```python
JB_Jul2021_df.to_pickle("Biden Speeches Raw/JB_Jul2021_df.pkl")
JB_Aug2021_df.to_pickle("Biden Speeches Raw/JB_Aug2021_df.pkl")
JB_Aug2021_2_df.to_pickle("Biden Speeches Raw/JB_Aug2021_2_df.pkl")
```
